```python
import matplotlib.pyplot as plt
import json

def load_json(filename):
    with open(filename, 'r') as f:
        data = json.load(f)
    return data
```

```python
#data = load_json(r"C:\Users\amitm\VibrationAnalysis\okay_data\sdata_okay1.json")
data = load_json(r"..\ml\combined_data_okay.json")
data_fault = load_json(r"..\ml\combined_data_fault.json")
```

```python
okay_count = 0
fault_count = 0

for i in data:
    okay_count+=1

for i in data_fault:
    fault_count+=1
print(okay_count, "," , fault_count)
```

```python
time = range(len(data))

# Extract x, y, z values
x_values = [item[0] for item in data]
y_values = [item[1] for item in data]
z_values = [item[2] for item in data]

#Values of faulty data
x_fault_values = [item[0] for item in data_fault]
y_fault_values = [item[1] for item in data_fault]
z_fault_values = [item[2] for item in data_fault]


# Plotting
plt.subplot(2,1,1)
plt.plot(time, x_values, label="X-axis")
plt.plot(time, y_values, label="Y-axis")
plt.plot(time, z_values, label="Z-axis")

plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Normal Data")
#plt.legend()
plt.grid(True)  # Add grid for better visualization


plt.subplot(2,1,2)
plt.plot(time, x_fault_values, label="X-axis-fault")
plt.plot(time, y_fault_values, label="Y-axis-fault")
plt.plot(time, z_fault_values, label="Z-axis-fault")

plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Faulty Data")
#plt.legend()
plt.grid(True)  # Add grid for better visualization


plt.tight_layout()
plt.show()
```
